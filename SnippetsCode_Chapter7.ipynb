```python
people = {
    "first": ["Corey", "Jane", "John", "Adam"],
    "last": ["Schafer", "Doe", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com', 'A@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

## Ordenação crescente

```python
df.sort_values(by='last')
```

## Ordenação decrescente

```python
df.sort_values(by='last', ascending=False)
```

## Ordenação por múltiplas colunas
A hierarquia da ordenação, segue da primeira para as seguintes colunas apresentadas pela lista no parâmetro `by=`

```python
df.sort_values(by=['last', 'first'], ascending=False)
```

## Como criar diferentes tipos de ordenação para diferentes colunas

```python
df.sort_values(by=['last', 'first'], ascending=[False, True])
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df
```

Para retornar ao dataframe original, basta solicitar uma ordenação apartir dos indices

```python
df.sort_index()
```

## Ordenação de uma série dentro de um dataframe

```python
df['last'].sort_values()
```
